```python
import pandas as pd 
```

```python
import requests

url = 'http://data.cityofnewyork.us/api/views/jb7j-dtam/rows.json'
results = requests.get(url).json() # reading in the json 
```

```python
results.keys()
```

```python
results['data']
```

```python
df = pd.DataFrame(results["data"])
df.head()
```

```python
columns = results["meta"]["view"]["columns"] # this gives us the descriptions and names for the columns
columns
```

```python
headers = [i["fieldName"] for i in columns] # we create a list of the column names by list comprehension
headers
```

```python
df = pd.DataFrame(results["data"], columns=headers) # and then pass in a list of those column names to our df
df
```

```python
df.drop(labels = [':sid', ':position', ':meta', ':created_meta', ':updated_meta'], axis=1, inplace=True)
df
```

```python
df.drop(df.index[1094:], inplace=True) #The last data three rows are metadata, so we drop them
df
```

```python
renaming_dict = {
    ':id': 'key', 
    ':created_at': 'created_at', 
    ':updated_at': 'updated_at'
}

df.rename(columns=renaming_dict, inplace=True)
df

#renames columns using a dictionary
```

```python
df["year"] = pd.to_numeric(df["year"])   #Changes datatype of year to an int64
df.dtypes
```

```python
#What was the leading cause of death in 2012?

df2012 = df[df['year'] == 2012]
df2012['leading_cause'].describe()[2]
```

```python
df2013 = df[df['year'] == 2013]
df2013['race_ethnicity'].unique()
```

```python
#What race/ethnicity had the most recorded deaths in 2013?
df2013['race_ethnicity'].describe()[2]
```

```python
df_not_dots = df[df['age_adjusted_death_rate'] != '.']  #removes entries with null or NaN values for age_adjusted_death_rate
df_not_dots 
```

```python
new_df = df_not_dots.sort_values(by="age_adjusted_death_rate",ascending=True) #Sorts the remaining entries by ascending age_adjusted_death_rate value
new_df
```

```python
#Which year had the lowest age adjusted death rate?
new_df.head(1)['year'] #Specifies that the year is 2014 and is the 953rd entry
```
